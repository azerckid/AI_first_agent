```python
import os
from dotenv import load_dotenv
import google.generativeai as genai
from google.generativeai.types import GenerationConfig

load_dotenv()

genai.configure(api_key=os.environ["GEMINI_API_KEY"])

model = genai.GenerativeModel(
    "gemini-2.0-flash",
    system_instruction="""
    You have the following functions in my system:
    `get_weather`
    `get_news`
    `get_currency`
    
    When the user asks a question, identify the appropriate function.
    Then, provide a detailed answer to the question.
    Finally, wrap your detailed answer inside the function call.
    
    Format: function_name(detailed_answer)
    
    Example:
    User: What is the weather in Spain?
    AI: get_weather(The weather in Spain is sunny with temperatures around 25 degrees...)
    """
)

chat = model.start_chat(history=[])

while True:
    user_question = input("Ask a question (or type 'q' to quit): ")
    if user_question.lower() in ["q", "quit"]:
        break      
    response = chat.send_message(user_question)
    print(f"AI: {response.text}")
```

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv()

genai.configure(api_key=os.environ["GEMINI_API_KEY"])

def get_weather(country):
    return f"The weather in {country} is sunny."

def get_news(country):
    return f"News in {country}: No major events today."

def get_currency(country):
    return f"Currency in {country} is Euro."

tools = [get_weather, get_news, get_currency]

model_with_tools = genai.GenerativeModel(
    "gemini-2.0-flash",
    tools=tools
)

chat_with_tools = model_with_tools.start_chat(enable_automatic_function_calling=True)

while True:
    user_question = input("Ask a question to Tool Agent (or type 'q' to quit): ")
    if user_question.lower() in ["q", "quit"]:
        break      
    response = chat_with_tools.send_message(user_question)
    print(f"Tool Agent: {response.text}")
```
